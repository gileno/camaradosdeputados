```python
%matplotlib inline
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pydataset
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_w = pd.read_csv('winequality-white.csv', delimiter=';')
```

```python
data_w.head()
```

```python
data_r = pd.read_csv('winequality-red.csv', delimiter=';')
```

```python
data_r.head()
```

```python
data_r['class'] = np.nan
```

```python
data_r.head()
```

```python
data_r['class'].fillna('red', inplace=True)
```

```python
data_w['class'] = np.nan
data_w['class'].fillna('white', inplace=True)
```

```python
data_w.head()
```

```python
data = data_w.append(data_r)
```

```python
data.head()
```

```python
data.tail()
```

```python
knn = KNeighborsClassifier(11)
x_treino, x_teste, y_treino, y_teste = train_test_split(
    data[['residual sugar', 'density', 'sulphates', 'alcohol', 'fixed acidity']],
    data['class'], test_size=0.3
)
```

```python
knn.fit(x_treino, y_treino)
```

```python
knn.score(x_teste, y_teste)
```

```python
resultados = {}
for k in range(1, 50, 5):
    knn = KNeighborsClassifier(k)
    x_treino, x_teste, y_treino, y_teste = train_test_split(
        data[['residual sugar', 'density', 'sulphates', 'fixed acidity']],
        data['class'], test_size=0.3
    )
    knn.fit(x_treino, y_treino)
    resultados[k] = knn.score(x_teste, y_teste)
```

```python
plt.plot(list(resultados.keys()), list(resultados.values()), 'b-')
```
