```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import matplotlib.dates as dates
import datetime as dt
```

```python
df = pd.read_csv('ppz-jan-fev-2017.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
def to_date(value):
    return dt.datetime(2017, 1, 1) + dt.timedelta(hours=value)
df['date'] = df['hour'].apply(to_date)
```

```python
df.head()
```

```python
del df['hour']
```

```python
df.tail()
```

```python
df.set_index(['date'], inplace=True)
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
ax.plot_date(df.index.to_pydatetime(), df['views'], 'b-')
ax.xaxis.set_minor_locator(dates.DayLocator(bymonthday=range(5, 32, 5)))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
ax.xaxis.grid(True, which="minor")
ax.yaxis.grid()
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b'))
plt.tight_layout()
plt.show()
```
