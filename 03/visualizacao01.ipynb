```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1, 10, 101)
```

```python
x
```

```python
plt.plot(x, np.sin(x))
plt.show()
```

```python
np.sin(1)
```

```python
plt.plot(x, np.sin(x), 'r')
plt.show()
```

```python
plt.plot(x, np.sin(x), '--r')
plt.show()
```

```python
plt.plot(x, np.sin(x), '--g')
plt.show()
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '--g')
plt.plot(x, np.cos(x), '--r')
plt.show()
```
