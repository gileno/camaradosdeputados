```python
pydataset.data('iris', show_doc=True)
```

```python
iris = pydataset.data('iris')
```

```python
iris.head()
```

```python
iris['Species'].unique()
```

```python
plt.scatter(
    iris['Sepal.Length'], iris['Sepal.Width'],
    sizes=50 * iris['Petal.Length']
)
```

```python
def specie_color(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    return 2

iris['SpecieColor'] = iris['Species'].apply(specie_color)
```

```python
iris.head()
```

```python
plt.scatter(
    iris['Sepal.Length'], iris['Sepal.Width'],
    sizes=50 * iris['Petal.Length'], cmap='viridis', alpha=0.4,
    c=iris['SpecieColor']
)
```
