```python
import pandas as pd
import numpy as np
```

```python
primary_df = pd.read_csv('primary_results.csv')
```

```python
primary_df['candidate'].unique()
```

```python
pd.pivot_table(
    primary_df, index=['state', 'party', 'candidate'], values=['votes'],
    aggfunc={'votes': np.sum}
)
```

```python
primary_df['rank'] = primary_df.groupby(
    ['county', 'party'])['votes'].rank(ascending=False)
```

```python
primary_df[primary_df['county'] == 'Los Angeles'][['candidate', 'rank']]
```
