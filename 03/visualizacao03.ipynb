```python
estados = pd.read_csv('populacao_brasil.csv')
```

```python
estados.head()
```

```python
estados['total'].hist()
```

```python
fig, ax = plt.subplots()
plt.hist(estados['total'], bins=15, orientation='horizontal')
ax.ticklabel_format(style='plain')
plt.show()
```

```python
estados['percent'] = estados['total'] / estados['total'].sum()
```

```python
estados.head()
```

```python
estados['total'].sum()
```

```python
plt.pie(estados['percent'], labels=estados['estado'])
plt.show()
```

```python
plt.pie(estados['percent'], labels=estados['estado'], autopct='%1.2f%%')
plt.show()
```

```python
estados_filtro = estados[estados['percent'] > 0.04]
plt.pie(
    estados_filtro['percent'], labels=estados_filtro['estado'],
    autopct='%1.2f%%'
)
plt.show()
```

```python
estados_f = estados[estados['percent'] > 0.04]
estados_f['percent'] = estados_f['percent'] / estados_f['percent'].sum()
plt.pie(
    estados_f['percent'], labels=estados_f['estado'],
    autopct='%1.2f%%'
)
plt.show()
```

```python
estados.sort_values('percent', ascending=False)
```
