```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pydataset

from sklearn.cluster import KMeans
```

```python
iris = pydataset.data('iris')
```

```python
iris.head()
```

```python
X = iris[['Sepal.Width', 'Sepal.Length', 'Petal.Width', 'Petal.Length']]
Y = iris['Species'].apply(lambda x: {'setosa': 0, 'virginica': 1, 'versicolor': 2}[x])
```

```python
colormap = np.array(['red', 'lime', 'black'])
```

```python
plt.figure(figsize=(18, 8))

plt.subplot(1, 2, 1)
plt.scatter(X['Petal.Length'], X['Petal.Width'], c=colormap[Y])
plt.title('Petal')

plt.subplot(1, 2, 2)
plt.scatter(X['Sepal.Length'], X['Sepal.Width'], c=colormap[Y])
plt.title('Sepal')
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
plt.figure(figsize=(18, 8))

plt.subplot(1, 2, 1)
plt.scatter(X['Petal.Length'], X['Petal.Width'], c=colormap[kmeans.labels_])
plt.title('Petal')

plt.subplot(1, 2, 2)
plt.scatter(X['Sepal.Length'], X['Sepal.Width'], c=colormap[kmeans.labels_])
plt.title('Sepal')
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

plt.figure(figsize=(18, 8))

plt.subplot(1, 2, 1)
plt.scatter(X['Petal.Length'], X['Petal.Width'], c=colormap[kmeans.labels_])
plt.title('Petal')

plt.subplot(1, 2, 2)
plt.scatter(X['Sepal.Length'], X['Sepal.Width'], c=colormap[kmeans.labels_])
plt.title('Sepal')
```

```python
colormap = np.array(['red', 'lime', 'black', 'white', 'blue'])

kmeans = KMeans(n_clusters=3)
kmeans.fit(X[['Petal.Length', 'Petal.Width']])

plt.figure(figsize=(18, 8))

plt.subplot(1, 2, 1)
plt.scatter(X['Petal.Length'], X['Petal.Width'], c=colormap[kmeans.labels_])
plt.title('Petal')

plt.subplot(1, 2, 2)
plt.scatter(X['Sepal.Length'], X['Sepal.Width'], c=colormap[kmeans.labels_])
plt.title('Sepal')
```

```python
colormap = np.array(['red', 'lime', 'black', 'white', 'blue'])

kmeans = KMeans(n_clusters=5)
kmeans.fit(X)

plt.figure(figsize=(18, 8))

plt.subplot(1, 2, 1)
plt.scatter(X['Petal.Length'], X['Petal.Width'], c=colormap[kmeans.labels_])
plt.title('Petal')

plt.subplot(1, 2, 2)
plt.scatter(X['Sepal.Length'], X['Sepal.Width'], c=colormap[kmeans.labels_])
plt.title('Sepal')
```
