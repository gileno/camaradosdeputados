```python
dataset = [
    ['leite', 'cebola', 'manteiga', 'feijão', 'ovos', 'iogurte'],
    ['sal', 'cebola', 'manteiga', 'feijão', 'ovos', 'iogurte'],
    ['leite', 'maça', 'feijão', 'ovos'],
    ['leite', 'arroz', 'pipoca', 'feijão', 'iogurte'],
    ['pipoca', 'cebola', 'feijão', 'sorvete', 'ovos']
]
```

```python
import pandas as pd
from mlxtend.preprocessing import OnehotTransactions
```

```python
oht = OnehotTransactions()
```

```python
oht_array = oht.fit(dataset).transform(dataset)
df = pd.DataFrame(oht_array, columns=oht.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.6)
```

```python
apriori(df, min_support=0.7, use_colnames=True)
```

```python
freq_items = apriori(df, min_support=0.6, use_colnames=True)
freq_items['tamanho'] = freq_items['itemsets'].apply(lambda x: len(x))
freq_items
```

```python
freq_items.columns = ['Suporte', 'ConjuntoItens', 'Tamanho']
```
